```python
# Bibliotecas necessárias ao script:
import numpy as np  # importa a biblioteca numpy (que trabalha com arrays numéricos)
import time         # importa a biblioteca utilizada para contar o tempo
import pandas as pd # biblioteca para trabalhar com data frames
import math
from numpy import random

################################################
# Algoritmos de ordenação
################################################

# Bubblesort
def bubble_sort(array): 
    trocas = comparacoes = pos_troca = 0
    qtd_elementos = len(array)-1
    troca = True
    
    while troca:
        troca = False
        for i in range(0, qtd_elementos):
            comparacoes = comparacoes + 1
            if array[i] > array[i+1]:
                tmp = array[i]
                array[i] = array[i+1]
                array[i+1] = tmp
                troca = True  
                pos_troca = i
                trocas = trocas + 1
        qtd_elementos = pos_troca
                
    return {'trocas':trocas, 'comparacoes':comparacoes}

# Quicksort e funções auxiliares
def quick_sort(array):
    log_operacoes = {'trocas':0, 'comparacoes':0}
    quicksort(array, 0, len(array)-1, log_operacoes)
    return log_operacoes

def quicksort(array, inicio, fim, log_operacoes):
    if fim > inicio:
        pivo = particiona(array, inicio, fim, log_operacoes)
        quicksort(array, inicio, pivo-1, log_operacoes)
        quicksort(array, pivo+1, fim, log_operacoes)
    return log_operacoes

def particiona(array, esquerda, direita, log_operacoes):
    pivo = esquerda
    i = pivo + 1 
    j = direita
    
    while i<j:
        while (array[j] >= array[pivo]) and (j > esquerda): # procura menor à direita
            j = j - 1
            log_operacoes['comparacoes'] = log_operacoes['comparacoes'] + 1 
        
        while (array[i] < array[pivo]) and (i < direita): # procura maior à esquerda
            log_operacoes['comparacoes'] = log_operacoes['comparacoes'] + 1 
            i = i + 1
            
        if (i<j) and (array[i] > array[j]):
            log_operacoes['comparacoes'] = log_operacoes['comparacoes'] + 1 
            log_operacoes['trocas'] = log_operacoes['trocas'] + 1 
            tmp = array[i]
            array[i] = array[j]
            array[j] = tmp

    if array[j] < array[pivo]:
        log_operacoes['comparacoes'] = log_operacoes['comparacoes'] + 1 
        log_operacoes['trocas'] = log_operacoes['trocas'] + 1 
        tmp = array[j]
        array[j] = array[pivo]
        array[pivo] = tmp
    
    return j

################################################
# Implementação dos seus algoritmos 

def comb_sort(array):
    log_operacoes = {'trocas':0, 'comparacoes':0}
    tam = len(array)  
    troca = True                    # inicializa como true pra rodar o primeiro laço
    gap = tam                       # gap inicial tem o tamanho do array

    # continua enquanto o gap for maior que 1 e ocorra troca na última comparação
    log_operacoes['comparacoes'] = log_operacoes['comparacoes'] + 1 
    while gap !=1 or troca:  
        # acha próximo gap 
        gap = int((gap * 10)/13)	# diminui o gap usando o shrink factor
        log_operacoes['comparacoes'] = log_operacoes['comparacoes'] + 1
        if gap < 1:                 # se for menor que 1, satura em 1
            gap = 1
        
        troca = False               # para saber se houve troca dentro do laço

        # compara cada elemento com o seu correspondente (pulando o gap) 
        for i in range(0, tam-gap):
            log_operacoes['comparacoes'] = log_operacoes['comparacoes'] + 1 
            if array[i] > array[i + gap]: 
                array[i], array[i + gap] = array[i + gap], array[i] 
                log_operacoes['trocas'] = log_operacoes['trocas'] + 1
                troca = True
                
    return log_operacoes 


def shake_sort(array):
    log_operacoes = {'trocas':0, 'comparacoes':0}
    tam = len(array)
    troca = True
    inicio = 0
    fim = tam-1
    while (troca):
        troca = False  # reseta a flag

        # começa escaneando do início até o fim, ordenando como o bubble sort
        for i in range(inicio, fim):
            log_operacoes['comparacoes'] = log_operacoes['comparacoes'] + 1 
            if (array[i] > array[i + 1]):
                array[i], array[i + 1] = array[i + 1], array[i]
                log_operacoes['trocas'] = log_operacoes['trocas'] + 1
                troca = True

        # se não houve troca, o array está ordenado, para o loop
        if (troca == False):
            break
        
        troca = False   # se não, reseta a flag pro próximo loop
        fim = fim-1     # pula o último elemento, que já está na posição correta
        
        # escaneia do fim ao início
        for i in range(fim-1, inicio-1, -1):
            log_operacoes['comparacoes'] = log_operacoes['comparacoes'] + 1 
            if (array[i] > array[i + 1]):
                array[i], array[i + 1] = array[i + 1], array[i]
                log_operacoes['trocas'] = log_operacoes['trocas'] + 1
                troca = True

        # o próximo scan começa um elemento à frente, pois o no início já está no lugar certo
        inicio = inicio + 1

    return log_operacoes 
   
# divide o array em 3 segmentos: menores, iguais e maiores que o pivô
def particiona2(array, inicio, fim, log_operacoes):
    # guarda os valores do primeiro elemento, da primeira e última posições, respectivamente
    pivo, j, t = array[inicio], inicio, fim
    # ponteiros i e j começam juntos
    i = j

    # enquanto i não tiver cruzado com o ponteiro t no fim do array, continua 
    while i <= t:
        log_operacoes['comparacoes'] = log_operacoes['comparacoes'] + 1
        # se o elemento na posição i for menor que pivô 
        if array[i] < pivo:
            log_operacoes['trocas'] = log_operacoes['trocas'] + 1
            # troca com o elemento apontado por j e avança j
            array[j], array[i] = array[i], array[j]
            j += 1

        # caso maior que o pivô
        elif array[i] > pivo:
            log_operacoes['trocas'] = log_operacoes['trocas'] + 1
            # troca ele com o elemento apontado por t e retrocede t
            array[t], array[i] = array[i], array[t]
            t -= 1
            i -= 1 # ponteiro i fica inalterado nesse caso
        
        # se eles forem iguais, o i só avança, sem modificar nada mais
        i += 1     # avança i
    # retorna j e t, os delimitadores da esquerda e da direita do segmento central    
    return j, t

def quickSort2(array, inicio, fim, log_operacoes):
    # caso array unitário
    if inicio >= fim:
        return

    # escolhe um pivô aleatório
    k = random.randint(inicio, fim)

    log_operacoes['trocas'] = log_operacoes['trocas'] + 1
    # troca ele de lugar com o elemento inicial
    array[inicio], array[k] = array[k], array[inicio]

    # captura delimitadores do segmento central (pivô e iguais a ele)
    m1, m2 = particiona2(array, inicio, fim, log_operacoes)

    # ordena segmentos à esquerda e à direita desse segmento central
    quickSort2(array, inicio, m1 - 1, log_operacoes)
    quickSort2(array, m2 + 1, fim, log_operacoes)

def quick_sort2(array):
    log_operacoes = {'trocas':0, 'comparacoes':0}
    quickSort2(array, 0, len(array)-1, log_operacoes)
    return log_operacoes


    # Fazer heapify com a subárvore de raíz no índice i. 
    # n é o tamanho do heap 
def heapify(arr, n, i, log_operacoes): 
    maior = i # Inicializa o maior elemento como raíz 
    esq = 2 * i + 1	 # esquerda = 2*i + 1 
    dir = 2 * i + 2	 # direita = 2*i + 2 

    # Verifica se filho esquerdo da raíz existe e se 
    # é maior que ela
    log_operacoes['comparacoes'] = log_operacoes['comparacoes'] + 1
    if esq < n and arr[i] < arr[esq]: 
        maior = esq 

    # Mesma coisa pro filho direito da raiz
    log_operacoes['comparacoes'] = log_operacoes['comparacoes'] + 1 
    if dir < n and arr[maior] < arr[dir]: 
        maior = dir 

    # Troca a raiz se precisar
    log_operacoes['comparacoes'] = log_operacoes['comparacoes'] + 1 
    if maior != i: 
        arr[i],arr[maior] = arr[maior],arr[i] 
        log_operacoes['trocas'] = log_operacoes['trocas'] + 1 
        # Faz Heapify na raiz 
        heapify(arr, n, maior, log_operacoes) 

def heapSort(arr):
    log_operacoes = {'trocas':0, 'comparacoes':0} 
    n = len(arr) 

    # Constrói o maxheap 
    # Já que o último nodo pai vai estar em ((n//2)-1), começar nesse índice 
    for i in range(n//2-1, -1, -1): 
        heapify(arr, n, i, log_operacoes) 

    # Extrai elementos um por um
    for i in range(n-1, 0, -1): 
        arr[i], arr[0] = arr[0], arr[i] # Troca
        log_operacoes['trocas'] = log_operacoes['trocas'] + 1 
        heapify(arr, i, 0, log_operacoes) 
    
    return log_operacoes
```
